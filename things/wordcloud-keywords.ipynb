```python
from wordcloud import WordCloud  #https://amueller.github.io/word_cloud
import matplotlib.pyplot as plt
```

```python
"""


WordCloud 

mask mask  mask 

colormap colormaphttps://matplotlib.org/3.1.0/tutorials/colors/colormaps.html colormap 'cubehelix'

WordCloudhttps://amueller.github.io/word_cloud/generated/wordcloud.WordCloud.html
"""

keywords = {'Python':2,'CodeLab Scratch':2, 'CodeLab Adapter':2, '':2, 'Linux':2, 'JupyterLab':2, 'Madmom':1, '':1, '':2, '':3, '':2, '':2,  
            'Beats':2, '':2, '':2, 'Inkscape':1, 'GIMP':1, 'Keras':1, '':2, '':2, '':1, 'fMRI':1 }
font_path = 'NotoSansCJK-Bold.ttc'            # 
#mask = np.array(Image.open('./img/fmask.png'))  #  mask 
keywords_wc = WordCloud(mask = None, width=1920, height=1080, font_path=font_path,  colormap='twilight', max_words=1000).generate_from_frequencies(keywords)  # 
   
plt.figure(figsize=(20,20))
plt.axis('off')
plt.imshow(keywords_wc,interpolation='bilinear')
plt.savefig('keywordss.png',dpi=300, facecolor='k', bbox_inches='tight')    # facecolorbbox_inchestightdpi
```
